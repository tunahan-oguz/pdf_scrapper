```python
def get_image_files(directory):
    image_files = []
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(('.png', '.jpeg', '.jpg')):
                image_files.append(os.path.join(root, file))
    return image_files
```

```python
import os
import cv2
import numpy as np
import keras_ocr
import math

def midpoint(x1, y1, x2, y2):
    return (int((x1 + x2) / 2), int((y1 + y2) / 2))

def inpaint_text(img_path, pipeline, output_path):
    img_path = img_path.replace('\\', '/')
    output_path = output_path.replace('\\', '/')

    img = keras_ocr.tools.read(img_path)
    if img is None:
        print(f"Error: Unable to read image from {img_path}")
        return

    # Convert from RGB to BGR
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

    prediction_groups = pipeline.recognize([img])
    if not prediction_groups[0]:
        print(f"Error: No text detected in the image.")
        success = cv2.imwrite(output_path, img)
        return

    mask = np.zeros(img.shape[:2], dtype="uint8")
    text = ""

    for box in prediction_groups[0]:
        text += f"{box[0]} "
        x0, y0 = box[1][0]
        x1, y1 = box[1][1]
        x2, y2 = box[1][2]
        x3, y3 = box[1][3]

        x_mid0, y_mid0 = midpoint(x1, y1, x2, y2)
        x_mid1, y_mid1 = midpoint(x0, y0, x3, y3)
        thickness = int(math.sqrt((x2 - x1)**2 + (y2 - y1)**2))

        cv2.line(mask, (x_mid0, y_mid0), (x_mid1, y_mid1), 255, thickness)

    inpainted_img = cv2.inpaint(img, mask, 7, cv2.INPAINT_NS)

    output_dir = os.path.dirname(output_path)
    if not os.path.exists(output_dir):
        try:
            os.makedirs(output_dir)
        except Exception as e:
            print(f"Error: Unable to create directory {output_dir}. Exception: {e}")
            return

    # Save the inpainted image
    success = cv2.imwrite(output_path, inpainted_img)
    if not success:
        print(f"Error: Unable to save image to {output_path}")
        return

    print(f"Image saved successfully to {output_path}")

    return [text.strip(), output_path]
```
