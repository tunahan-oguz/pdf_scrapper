```python
import pandas as pd 
import os 
import glob
from gensim.models import Word2Vec
```

```python
csv_files = glob.glob('dataset/descriptions/*.csv')
csv_dfs= []
for file in csv_files:
    csv_dfs.append(pd.read_csv(file))

```

```python
descriptions = []
for csv in csv_dfs:
    descriptions.extend(csv['Description_Alphanumeric'].tolist())
descriptions = [s.split() for s in descriptions]
```

```python
descriptions
```

```python
model = Word2Vec(sentences=descriptions, seed=0, workers=1, sg=0)
```

```python
similar_words = model.wv.most_similar('temporal', topn=5)
```
